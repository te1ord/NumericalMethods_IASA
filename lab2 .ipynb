 ## Знайти розв’язки перетвореної системи  точними методами Крамера, оберненої матриці, LU – декомпозиції.

```python
def cramer(A, b):
    n = A.shape[0]
    det_A = A.det()
    x = []
    for i in range(n):
        matrix = A.copy()
        matrix[:,i] = b
        det_i = matrix.det()
        x = det_i / det_A
        x.append(x.evalf())
    return x
```

```python
def LU_decomposition(A, b):
    n = A.shape[0]
    L, U, _ = A.LUdecomposition()
    y = []
    for i in range(n):
        y_i = b[i]
        for j in range(i):
            y_i -= L[i, j]*y[j]
        y.append(y_i)
    x = []
    for i in range(n-1, -1, -1):
        x_i = y[i]
        for j in range(i+1, n):
            x_i -= U[i, j]*x[n-j-1]
        x_i /= U[i, i]
        x.append(x_i)
    x.reverse()
    return x
```

## Скласти програми ітераційних методів Якобі, Зейделя. Провести серію обчислювальних експериментів із варіацією значень ε (ε = 0.1, ε = 0.01, ε = 0.001, ε = 10−6) і початкових наближень. Визначити залежності, які впливають на кількість ітерацій.

```python
x0 = [0, 0, 0, 0, 0, 0]
epsilons = [0.1, 0.01, 0.001, 0.000001]
```

```python
def jacobi_method(A, b, x0, epsilon):
    n = A.shape[0]
    x = x0.copy()
    iterations = 0
    while True:
        iterations += 1
        x_new = []
        for i in range(n):
            s = 0
            for j in range(n):
                if j != i:
                    s += A[i, j] * x[j]
            x_i = (b[i] - s) / A[i, i]
            x_new.append(x_i)
        if max([abs(x_new[i]-x[i]) for i in range(n)]) < epsilon:
            break
        x = x_new
    return x, iterations
```

### Розв'язки методом Якобі:

```python
for epsilon in epsilons:
    print(f"epsilon={epsilon}")
    x_jacobi, iterations_jacobi = jacobi_method(a_vis, b, x0, epsilon)
    print(f"Метод Якобі: x={x_jacobi} кількість ітерацій={iterations_jacobi}")
```

```python
def zeidel(A, b, x0, epsilon):
    n = A.shape[0]
    x = x0.copy()
    iterations = 0
    while True:
        iterations += 1
        for i in range(n):
            s = 0
            for j in range(n):
                if j != i:
                    s += A[i, j] * x[j]
            x[i] = (b[i] - s) / A[i, i]
        if max([abs(x[i]-x0[i]) for i in range(n)]) < epsilon:
            break
        x0 = x.copy()
    return x, iterations
```

### Бачимо, що наша матриця не симетрична, тож маємо провести її симетризацію задля забезпечення збіжності.

```python
A = (a_vis + a_vis.T)/2
A.is_symmetric()
```

```python
def upper_relaxation_method(A, b, w, x0, epsilon, max_iter):
    n = len(x0)
    x = x0.copy()
    num_iters = 0
    for _ in range(max_iter):
        num_iters += 1
        x_new = x.copy()
        for i in range(n):
            s = sum(-A[i, j] * x_new[j] for j in range(n) if i != j)
            x_new[i] = (1 - w) * x[i] + w * (b[i] + s) / A[i, i]
        if max(abs(x_new[i] - x[i]) for i in range(n)) < epsilon:
            return x_new, num_iters
        x = x_new
    return x, num_iters
```

### Ров'язок методом верхньої релаксації для w = 0.1, epsilon = 0.1, max_iter = 100(як приклад)

```python
upper_relaxation_X = upper_relaxation_method(A, b, 0.1, x0, 0.1, 100)
Matrix(upper_relaxation_X[0])
```

### Тепер визначимо оптимальні значення параметрів epsilon та omega.

```python
omegas = [0.1, 0.5, 0.9, 1.1, 1.5, 1.9] 
max_iter = 1000
```

```python
optimal_omega = None
optimal_epsilon = None
min_iterations = float("inf")


for epsilon in epsilons:
    print(f"Точність: {epsilon}")
    for omega in omegas:
        x, iters = upper_relaxation_method(A, b, omega, x0, epsilon, max_iter)
        print(f"omega = {omega}, розв'язок: {x}, кількість ітерацій: {iters}")
        
        if iters < min_iterations:
            min_iterations = iters
            optimal_omega = omega
            optimal_epsilon = epsilon


print(f"Оптимальний варіант: omega = {optimal_omega}, точність = {optimal_epsilon}, кількість ітерацій: {min_iterations}")
```
